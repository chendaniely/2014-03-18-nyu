Linear regression with one feature
We are interested in studying the relationship between age and height (statures) in girls aged 2 to 8
years old. We think that this can be modeled with a linear regression model. We have examples (data
points) for a population of 60 girls. Each example has one feature Age along with a numerical label
Height. We will use the dataset girls train.csv (based on real data from CDC). Your mission is to
implement linear regression with one feature using gradient descent. You will plot the data, regression
line, coefficient contours and cost function. You will finally make a prediction for a new example using
your regression model and assess your out of sample mean square error.

```python
# load train.csv into pandas dataframe
df = pd.read_csv('girls_train.csv', header = None)

# print the first 5 rows
print df[:5]
```

```python
# rename columns in the dataframe
df.columns = ['age', 'height',2,3]
```

#### b) Plot the distribution of the data. You should be getting a plot similar to Figure 1.

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.scatter(df['age'], df['height'])
```

```python
from ggplot import *

ggplot(aes(x='age', y='height'), data=df) + geom_point()
```

```python
import statsmodels.api as sm

regr_results = sm.OLS.from_formula('height ~ age', df).fit()
regr_results.summary()
```

```python
print regr_results.params
print regr_results.rsquared
```

```python
parameters = regr_results.params
rsquared = regr_results.rsquared
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

predicted = parameters[0] + parameters[1] * df['age']
plt.plot(df['age'], df['height'], 'ro')

#min_age, max_age = min(df['height']), max(df['height'])
plt.plot(df['age'], predicted, 'k-')
```

```python
min_age, max_age = min(df['height']), max(df['height'])
plt.plot([min_age, max_age], [min_age, max_age], 'k-')
```
